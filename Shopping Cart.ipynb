```python
def add_item(a_list, item_dict):
    return a_list.append(item_dict)

def remove_item(a_list, item_dict):
    for p in a_list:
        if p['name'] == item_dict['name']:
            a_list.remove(p)
            
def show_items(a_list):
    total = 0
    print('=' * 40)
    if not a_list:
        print("Your cart is empty. ")
    else:
        display_cart = []
        for p in a_list:
            if p['name'] not in [i['name'] for i in display_cart]:
                display_cart.append({
                    'name': p['name'],
                    'price': p['price']
                })
            else:
                for i in display_cart:
                    if i['name'] == p['name']:
                        i['price'] += p['price']
                        
        for i in range(len(display_cart)):
            print(f"{i+1}: {display_cart[i]['name']} [{[i['name'] for i in a_list].count(display_cart[i]['name'])}] ${display_cart}")
            
    print('=' * 40)
    
    total += float(sum([p['price'] for p in a_list]))
    print(f"total: ${total:,.2f}")
    
def clear_items(a_list):
    a_list.clear()
    
def main():
    cart = []
    
    done= False
    while not done:
        show_items(cart)
        ask = input("What would you like to do? ")
        
        if ask == "add":
            cart_name = input("What would you like to add? ")
            cart_price = float(input("How much does it cost? "))
            cart_item = {
                'name': cart_name,
                'price': cart_price
            }
            add_item(cart, cart_item)
        elif ask == 'show':
            show_items(cart)
            confirm = input("Press 'ENTER' when you're done viewing your cart. ")
        elif ask == "remove":
            cart_item = input("What would you like to remove? ")
            for item in cart:
                if item["name"] == cart_item:
                    remove_item(cart, item)
        elif ask == "clear":
            clear_items(cart)
        elif ask == "quit":
            confirm = input("Are you sure? Y/N ")
            if confirm == "y":
                done = True
            elif confirm == "n":
                continue
                
    return

main()


                
                    
                    
    
```
